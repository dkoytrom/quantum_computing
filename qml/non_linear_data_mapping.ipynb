# Non linear data mapping to higher dimension

In this demonstration we will use the circles dataset (https://scikit-learn.org/stable/modules/generated/sklearn.datasets.make_circles.html), a toy dataset for classification algorithms

```python
import matplotlib.pyplot as plt
import numpy as np
import pylab as pl
from sklearn.datasets import make_circles

# we have to generate our input dataset
X, Y = make_circles(n_samples = 200, noise = 0.05, factor = 0.4)

# we can seperate the circles based on their labels
A = X[np.where(Y == 0)]
B = X[np.where(Y == 1)]

# plot data in 2d plane, as we have 2 features in our dataset
plt.figure(figsize = (5, 5))

plt.scatter(A[:, 0], A[:, 1], marker = 'o')
plt.scatter(B[:, 0], B[:, 1], marker = 'x', c = 'C3')

plt.show()
```

* Our dataset has a pattern, but data are not linearly seperable in 2D. Let's map out data in 3D to see what happens

Extra dimension wil be calculated as 
$$ z = x^2 + y^2 $$

```python
def map_data(x, y):
    return np.array([x, y, x**2.0 + y**2.0])

tmp = []

for x, y in zip(np.ravel(A[: ,0]), np.ravel(A[:, 1])) :
    tmp.append(map_data(x, y))

A_3d = np.array(tmp)

tmp = []

for x, y in zip(np.ravel(B[: ,0]), np.ravel(B[:, 1])) :
    tmp.append(map_data(x, y))

B_3d = np.array(tmp)

# after mapping to 3d, we can plot the new 3d data
fig = plt.figure(figsize = (10, 5))

axes = plt.axes(projection='3d')

axes.scatter(A_3d[:, 0], A_3d[:, 1], A_3d[:, 2], marker = 'o')
axes.scatter(B_3d[:, 0], B_3d[:, 1], B_3d[:, 2], marker = 'x', c = 'C3')

axes.view_init(5, 60)

fig.show()
```

After mapping, pattern on data is more clear, so we can use a hyperplane to separate the data, annd it is clear that the plane is not linear in 2d
